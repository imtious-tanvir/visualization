```python
mu = 5
k = 2
x= np.arange(0, 10)
p = poisson.pmf(x , mu)
```

```python
plt.bar(x, p, color='mediumseagreen', edgecolor='black')
plt.bar(4, poisson.pmf(4, mu), color='orange', edgecolor='black')
plt.title("Poisson Distribution (Î» = 5, K  = 4)", fontsize=14)
plt.xlabel("Number of Events (k)", fontsize=12)
plt.ylabel("Probability", fontsize=12)
plt.show()
```
