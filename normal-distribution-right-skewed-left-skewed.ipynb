```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

# Create figure
plt.figure(figsize=(12, 6))
```

```python

# 1️⃣ Normal Distribution
data_normal = np.random.normal(loc=50, scale=10, size=1000)
sns.histplot(data_normal, bins=20, kde=True, color='skyblue')
plt.title("Normal Distribution\n(Symmetric)")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
# 2️⃣ Right-Skewed Distribution
data_right = np.random.exponential(scale=10, size=1000)
sns.histplot(data_right, bins=20, kde=True, color='orange')
plt.title("Right-Skewed (Positively Skewed)")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
# Using a Beta Distribution with alpha > beta
data_left = np.random.beta(a=5, b=2, size=1000) * 100  # scale for visualization
sns.histplot(data_left, bins=20, kde=True, color='lightgreen')
plt.title("Left-Skewed (Negatively Skewed)")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```
