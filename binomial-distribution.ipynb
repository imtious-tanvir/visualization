```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom
```

```python
n = 5
p = 0.8
x = np.arange(0, n +1 )
y = binom.pmf(x, n, p)
```

```python
plt.bar(x, y, color='skyblue', edgecolor='black')
highlight_k = 4
plt.bar(highlight_k, binom.pmf(highlight_k, n, p), color='orange', edgecolor='black')
plt.title("Binomial Distribution: Exam Success (n=5, p=0.8)", fontsize=14)
plt.xlabel("Number of Exams Passed (k)", fontsize=12)
plt.ylabel("Probability", fontsize=12)
plt.show()
```
