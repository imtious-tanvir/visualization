```python
import numpy as np
from scipy.stats import multinomial
```

```python
# total trials
n = 10
```

```python
# probabilities for each face
p = [1/6] * 6   # equal probabilities
p
```

```python
# observed counts
x = [2, 3, 1, 1, 2, 1]
```

```python
# calculate probability using multinomial PMF
prob = multinomial.pmf(x, n=n, p=p)

print(f"Probability = {prob:.4f}")
```
