```python
import numpy as np
from scipy.stats import chi2
```

```python
# degrees of freedom
k = 5  
```

```python
# generate random chi-square distributed values
data = chi2.rvs(df=k, size=1000)
```

```python
# calculate probability for a given value
x = 10
p = chi2.cdf(x, df=k)
print(f"P(X <= {x}) =", p)
```
